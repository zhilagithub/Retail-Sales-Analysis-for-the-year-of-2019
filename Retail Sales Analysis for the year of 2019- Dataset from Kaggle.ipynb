```python
### Question 1: what was the best month for sales? how much was earned that month ? 


### Question 2: what city had the highest number of sales?





import pandas as pd

import matplotlib.pyplot as plt


# List of file names for the 12 CSV files
file_names = [
    "Sales_April_2019.csv",
    "Sales_August_2019.csv",
    "Sales_December_2019.csv",
    "Sales_February_2019.csv",
    "Sales_January_2019.csv",
    "Sales_July_2019.csv",
    "Sales_June_2019.csv",
    "Sales_March_2019.csv",
    "Sales_May_2019.csv",
    "Sales_November_2019.csv",
    "Sales_October_2019.csv",
    "Sales_September_2019.csv"
]

# Create an empty DataFrame to store the combined data
combined_data = pd.DataFrame()

# Loop through the list of file names and append their data to the combined_data DataFrame
for file in file_names:
    data = pd.read_csv(file)
    combined_data = combined_data.append(data, ignore_index=True)

# Save the combined data to a new CSV file
combined_data.to_csv('combined_sales_data.csv', index=False)
```

```python
### Question 1: what was the best month for sales? how much was earned that month ? 
```

```python
# add a column named "month" and populate it with information from the "Order Date" column

combined_data["Month"]=combined_data["Order Date"].str[0:2]


combined_data.head()
```

```python
combined_data.tail()
```

```python
# Add a 'City' column to the DataFrame and populate it with information from the 'Purchase Address' column.



#It splits the "Purchase Address" column using a comma (',') as the delimiter,
#and then selects the second part (index 1) of the split, removing any leading or trailing spaces using .str.strip().

combined_data["City"] = combined_data["Purchase Address"].str.split(',').str[1].str.strip()


combined_data.head()
```

```python

combined_data = combined_data.dropna()

combined_data
```

```python

combined_data = combined_data.drop_duplicates()

combined_data 
```

```python
# convert columns to correct data type



# Convert the 'Quantity Ordered' and 'Price Each' columns to numeric 

combined_data["Quantity Ordered"] = pd.to_numeric(combined_data["Quantity Ordered"], errors='coerce')

combined_data["Price Each"] = pd.to_numeric(combined_data["Price Each"], errors='coerce')

# Calculate the total sales for each row and store it in a new column 'sales'

combined_data["sales"] = combined_data["Quantity Ordered"] * combined_data["Price Each"]

# Now, you can access the 'sales' column to see the total sales for each row

combined_data["sales"]
```

```python
combined_data=combined_data.groupby("Month").sum()
```
